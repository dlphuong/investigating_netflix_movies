```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("dataset/netflix_data.csv")

movies = data[data["type"] == "Movie"]
movies_1990s = movies[
    (movies["release_year"] >= 1990) & (movies["release_year"] < 2000)
]

plt.hist(movies_1990s["duration"])
plt.title("Distribution of Movie Duration in the 1990s")
plt.xlabel("Duration (minute)")
plt.ylabel("Number of Movies")
plt.show()

duration = 100

short_movie_count = 0

action_movies_1990s = movies_1990s[movies_1990s["genre"] == "Action"]
for lab, row in action_movies_1990s.iterrows():
    if row["duration"] < 90:
        short_movie_count = short_movie_count + 1
    else:
        short_movie_count = short_movie_count
print(short_movie_count)
```
